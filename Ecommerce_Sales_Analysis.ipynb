```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")

print('\n Task 1: Basic Grouping and Single Aggregations\n')

Total_sales = df.groupby('region')['sales_amount'].sum()
print('\n Total_sales',Total_sales)

Number_transactions = df.groupby('category')['transaction_id'].count()
print('\n Number_transactions \n',Number_transactions)

Average_sales = df.groupby('salesperson')['sales_amount'].mean()
print('\n Average_sales \n',Average_sales)

Total_sales_new = df.groupby('region')['sales_amount'].sum().reset_index()
print('\n Total_sales_new \n',Total_sales_new)

Number_transactions_new = df.groupby('category')['transaction_id'].count().reset_index()
print('\n Number_transactions_new \n',Number_transactions_new)

Average_sales_new = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print('\n Average_sales_new \n',Average_sales_new)

regional_sales = Total_sales.sort_values(ascending=False).reset_index()
print('\n regional_sales \n',regional_sales)

print('\n Task 2: Multi-Column Grouping and Multiple Aggregations \n')

Total_sales_adv=df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print('\n Total_sales_adv \n',Total_sales_adv)

Sales_person_data=df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count'])
print ('\n Sales_person_data \n',Sales_person_data)

Sales_person_desc=Sales_person_data.sort_values(by='sum', ascending=False).reset_index()
print ('\n Sales_person_performance \n',Sales_person_desc)

total_sales_category = df.groupby('category')['sales_amount'].sum()
print('\n total_sales_category \n',total_sales_category)
top_category = total_sales_category.idxmax()
print('\n high generating revenue category \t:',top_category)

print( '\n Task 3: Custom Aggregation and Complete Sales Report \n')

def sales_range(x):
  return x.max() - x.min()

sales_range(df.sales_amount)
print('\n sales_range \n',sales_range(df.sales_amount))

region_analysis=df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print('\n region_analysis \n',region_analysis)

final_summary=df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
print('\n final_summary \n',final_summary)
```
