```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(0)
n = 200

df = pd.DataFrame({
    'age':      np.random.randint(18, 60, n),
    'income':   np.random.randint(20000, 120000, n),
    'spend':    np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})

#Task 1 — Bar Chart: Average Spend by Category
avg_spend = df.groupby('category')['spend'].mean()
avg_spend.plot(kind='bar', color='goldenrod', edgecolor='black',label= 'Average spend by category')

plt.xlabel('Spend')
plt.ylabel('Category')
plt.title('Average Spend by Category')
plt.legend()
plt.show()

print("\n Electronics category has the highest spend \n")

# Task 2 — Scatter Plot: Income vs Spend
sns.scatterplot(data=df, x='income', y='spend',color="steelblue")
plt.title('Income vs Spend')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.show()
print("\n There is no visible relationship or correlation between income and spend, as the points are distributed randomly across the plot.\n")

#Task 3 — Interactive Scatter Plot with Plotly
import plotly.express as px
fig=px.scatter(df,x='income',y='spend',color='category',title ='Income vs Expensiture',hover_data=['category'])
fig.show()
print ("\n The interactivity allows you to hover over specific points to see exact values and toggle categories on or off to isolate specific groups, which helps identify data that might be overlapping in a static view. \n")
```
